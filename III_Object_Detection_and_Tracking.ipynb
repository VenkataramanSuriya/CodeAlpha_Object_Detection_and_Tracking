# **III - Object Detection and Tracking**

**Install required libraries**

```python
!pip install ultralytics --quiet
!pip install opencv-python-headless --quiet

import cv2
from ultralytics import YOLO
```

**Load the YOLOv5 model**

```python
model = YOLO('/content/yolov5su.pt')
```

**Define the video file path**

```python
video_path = '/content/car-detection.mp4'
```

**Open the video file / Loop through each frame in the video**

```python
cap = cv2.VideoCapture(video_path)

ret, frame = cap.read()
if not ret:
    print("Failed to read the video")
else:

    out = cv2.VideoWriter('output_video.mp4', cv2.VideoWriter_fourcc(*'MP4V'), 20, (frame.shape[1], frame.shape[0]))

    while cap.isOpened():
        if not ret:
            break

        results = model(frame)

        for result in results:
          boxes = result.boxes
          for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0].int().tolist()
            label = model.names[int(box.cls)]
            confidence = box.conf.item()
            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
            cv2.putText(frame, f'{label} {confidence:.2f}', (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)


        out.write(frame)


        ret, frame = cap.read()

    # Release the VideoCapture and VideoWriter objects

    cap.release()
    out.release()

    print("Object detection and video processing completed!")

    # Download the processed video

    from google.colab import files
    files.download('output_video.mp4')
```
